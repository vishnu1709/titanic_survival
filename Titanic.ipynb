```python
import pandas as pd
import numpy as np
df=pd.read_csv("titanic.csv")
df.info()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent_1 = df.isnull().sum()/df.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
missing_data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.Sex=le.fit_transform(dfle.Sex)
dfle.Embarked=df.fillna('S')
dfle.Embarked=dfle.Embarked.astype(str)
dfle.Embarked=le.fit_transform(dfle.Embarked)
dfle.info()
```

```python
df.head()
```

```python
dfle.Ticket=le.fit_transform(dfle.Ticket)
```

```python
dfle.info()
```

```python
dfle.head()
dfle=dfle.drop(df[['PassengerId','Name','SibSp','Parch','Fare','Cabin']],axis="columns")
```

```python
dfle.Age=dfle.Age.fillna(dfle['Age'].mean())
dfle.head(60)
```

```python
#dfle['Embarked'] = dfle['Embarked'].map({'3+': 3})
```

```python
from sklearn.model_selection import train_test_split
X=dfle[['Pclass','Sex','Age','Ticket','Embarked']]
y=dfle['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
X_train,X_test,Y_train,Y_Test=train_test_split(X,y,test_size=0.2,random_state=0)
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
#from sklearn.linear_model import LogisticRegression
#LogReg=LogisticRegression()
#LogReg.fit(X_train,Y_train)
```

```python
Y_prediction = random_forest.predict(X_test)
Y_prediction
```

```python
YA=Y_Test.as_matrix(columns=None)
YA=YA.reshape((-1,1))
#Y_prediction=Y_prediction.as_matrix(columns=None)
Y_prediction=Y_prediction.reshape((-1,1))
```

```python
random_forest.score(X_test,YA)
```

```python
#using Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
LogReg=LogisticRegression()
LogReg.fit(X_train,Y_train)
Y_prediction=LogReg.predict(X_test)
from sklearn import metrics
confusionMatrix=metrics.confusion_matrix(Y_prediction,YA)
confusionMatrix
```

```python
print("Accuracy:",metrics.accuracy_score(Y_prediction,YA))
print("precision:",metrics.precision_score(Y_prediction,YA))
print("recall:",metrics.recall_score(Y_prediction,YA))
```

```python
#k nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(X_train,Y_train)
knn.score(X_test,YA)
```

```python
#using naive bayes
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train) 
Y_pred = gaussian.predict(X_test)
gaussian.score(X_test,YA)
```

```python
#using svm
from sklearn import svm
clf=svm.LinearSVC(loss='hinge')
clf.fit(X_train,Y_train)
clf.score(X_test,YA)
```

```python
#Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier()
decision_tree.fit(X_train,Y_train)
decision_tree.score(X_test,YA)
```
